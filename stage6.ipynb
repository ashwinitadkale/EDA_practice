```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\ashwi\Downloads\titanic\train.csv")
df.head()
```

```python
df['Age'].hist(bins=30, color='teal', edgecolor='black')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df, palette='pastel')
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare Colored by Survival")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,10))

sns.countplot(x='Sex', hue='Survived', data=df, ax=axes[0,0], palette='Set2')
sns.countplot(x='Pclass', hue='Survived', data=df, ax=axes[0,1])
sns.histplot(df['Age'], kde=True, color='skyblue', ax=axes[1,0])
sns.boxplot(x='Survived', y='Fare', data=df, ax=axes[1,1])

axes[0,0].set_title('Survival by Gender')
axes[0,1].set_title('Survival by Class')
axes[1,0].set_title('Age Distribution')
axes[1,1].set_title('Fare Distribution by Survival')

plt.tight_layout()
plt.show()
```
