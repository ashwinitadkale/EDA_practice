Convert Date Columns

```python
orders['order_purchase_timestamp']=pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at']=pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
```

```python
orders['delivery_time']=(
    orders['order_delivered_customer_date']-orders['order_purchase_timestamp']
)
```

```python
orders['delivery_time_days'] = orders['delivery_time'].dt.days
```

```python
orders['delivery_delay']=orders['order_estimated_delivery_date']-orders['order_delivered_customer_date']
```

```python
orders['delivery_delay_days']=orders['delivery_delay'].dt.days
orders['delivery_delay_days']
```

```python
delivered_orders = orders[orders['order_status'] == 'delivered']
```

```python
delivered_orders = delivered_orders[
    (delivered_orders['delivery_time_days'] >= 0) &
    (delivered_orders['delivery_time_days'] <= 60)
]
```

```python
delivered_orders['delivery_time_days'].describe()
```

REVIEW ANALYSIS

```python
reviews=pd.read_csv(r"C:\Users\ashwi\Desktop\practice\archive\olist_order_reviews_dataset.csv")
reviews.head()
```

```python
merged=pd.merge(delivered_orders,reviews,on='order_id')
```

Time Aggregation

```python
orders_payments['year']=orders_payments['order_purchase_timestamp'].dt.year
orders_payments['month']=orders_payments['order_purchase_timestamp'].dt.month
orders_payments['year_month']=orders_payments['order_purchase_timestamp'].dt.to_period('M')
```

```python
monthly_revenue=orders_payments.groupby('year_month')['payment_value'].sum()
monthly_revenu=monthly_revenue.sort_index()
monthly_revenu
```

```python
monthly_revenu.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Time")
plt.ylabel("Revenue")
plt.show()
```
