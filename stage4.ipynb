```python
import pandas as pd
matches=pd.read_csv("matches.csv")
deliveries=pd.read_csv("deliveries.csv")
matches.__dataframe__
print(matches.head())
print(deliveries.head())
```

```python
import matplotlib.pyplot as plt 
wins=matches.groupby(['season','winner']).size().unstack(fill_value=0)
wins.plot(kind='bar',stacked=True,figsize=(12,6))
plt.title("Matches won by teams in each season")
plt.show()
```

Toss decision v/s Match winner

```python
import seaborn as sns
sns.countplot(data=matches, x="toss_decision", hue="winner")
plt.title("Toss Decision vs Match Outcome")
plt.xticks(rotation=45)
plt.show()
```

Feature Engineering

```python
# Add a run rate column
deliveries['run_rate'] = deliveries['total_runs'] * 6 / (deliveries['over']+1)

print(deliveries[['match_id','over','total_runs','run_rate']].head())
```
