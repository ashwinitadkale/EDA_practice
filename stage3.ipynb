```python
import pandas as pd

df = pd.read_csv(r"C:\Users\ashwi\Desktop\practice\zomato.csv\zomato.csv", encoding="latin-1")

print(df.head())
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Cuisine distribution
top_cuisines = df['cuisines'].value_counts().head(10)
sns.barplot(x=top_cuisines.values, y=top_cuisines.index)
plt.title("Top 10 Cuisines in Bangalore")
plt.show()

# Price range distribution
sns.countplot(data=df, x='priceRange')
plt.title("Price Range Distribution")
plt.show()

# Ratings distribution
sns.histplot(df['aggregate_rating'], bins=20, kde=True)
plt.title("Restaurant Ratings Distribution")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# df = pd.read_csv("zomato.csv", encoding="latin-1")
# Ratings vs Price Range
sns.boxplot(x='approx_cost(for two people)', y='rate', data=df)
plt.title("Ratings vs Price Range")
plt.show()

# Online Order vs Ratings
sns.boxplot(x='online_order', y='rate', data=df)
plt.title("Do Restaurants with Online Orders Have Higher Ratings?")
plt.show()

# Table booking vs Rating
sns.boxplot(x='book_table', y='aggregate_rating', data=df)
plt.title("Do Restaurants with Table Booking Have Higher Ratings?")
plt.show()

# Correlation heatmap (numerical features only)
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
